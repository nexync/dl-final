```python
from __future__ import annotations

from minigrid.manual_control import ManualControl
import numpy as np
import random

from minigrid.wrappers import ImgObsWrapper
from gymnasium.core import ObservationWrapper
from stable_baselines3 import PPO
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor
from stable_baselines3.common.callbacks import BaseCallback

from stable_baselines3 import PPO
from stable_baselines3.common.policies import ActorCriticPolicy

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

from plot import make_plot
import matplotlib
matplotlib.use('TkAgg') 
import matplotlib.pyplot as plt
import time

from customenv import CustomDoorKey
from customfeatureextractor import CNNFeaturesExtractor
import gymnasium
import gym
from callback import CustomRewardCallback
```

```python
policy_kwargs = dict(
    features_extractor_class=CNNFeaturesExtractor,
    features_extractor_kwargs=dict(features_dim=128, regularization=True),
)
```

```python

# Setup model and environment
env = CustomDoorKey(size=5,  render_mode="rbg_array")
env = ImgObsWrapper(env)
model = PPO("CnnPolicy", env, policy_kwargs=policy_kwargs, verbose=1)

# Instantiate the callback
max_reward = 0.9
callback = CustomRewardCallback(check_freq=1000, reward_threshold=max_reward)

# Train the model
model.learn(total_timesteps=int(2e6), callback=callback)
model.save("model_5x5s")

# Optionally save the final plot
plt.savefig('5x5s.png')

# Close the environment
env.close()
```
